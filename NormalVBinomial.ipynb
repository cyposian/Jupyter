```python
#Visually Demonstrate (use histograms) to show that the Binomial and Normal distributions converge in the large N limit.
```

```python
import matplotlib.pyplot as plt
import numpy as np

n, p, size = 100000, .42, 100000
data = np.random.binomial(n, p, size)

plt.hist(data, bins=30, density='True', color='lightblue', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution at High N')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(0, 1, 100000)

plt.hist(data, bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution at High N')
plt.show()
```
